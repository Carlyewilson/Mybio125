```python
# import packages
import statistics
import pandas as pd
import scipy.stats as stats
from scipy.stats import sem
from statistics import mean, stdev
import matplotlib.pyplot as plt
import numpy as np
#include if using a jupyter notebook, remove if using a .py file
%matplotlib inline
```

```python
f_vc = [2.7, 3.7, 3.2, 3.45, 3.2, 3.1]
```

```python
m_vc = [3.3, 4.5, 3.7]
```

```python
f_vc, m_vc
```

```python
fvc_mean = mean(f_vc)
mvc_mean = mean(m_vc)
fvc_mean, mvc_mean
```

```python
f_vcsem = sem(f_vc)
m_vcsem = sem(m_vc)
```

```python
f_vcsem, m_vcsem
```

```python
plt.bar(['Female','Male'], [fvc_mean,mvc_mean], yerr=[fvc_sem, mvc_sem], capsize=10)
plt.xlabel('Biological Sex')
plt.ylabel('Vital Capacity (L)')
plt.title('Vital Capacity Differences in Females vs Males')
```
